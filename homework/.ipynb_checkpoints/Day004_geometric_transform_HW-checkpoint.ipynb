# 作業

實作本篇提到的三大概念

- 翻轉：實作上下左右的翻轉
- 縮放：比較鄰近差值與雙立方插值 (或雙線性插值) 的圖片品質
- 平移：建立 Translation Transformation Matrix 來做平移

```python
import cv2
import time
import numpy as np
import matplotlib.pyplot as plt

img_path = 'data/lena.png'
img = cv2.imread(img_path)
```

```python
# 水平翻轉 (horizontal)
img_hflip = img[:,::-1,:]
cv2.imshow('img_hflip', img_hflip)
plt.figure(figsize=(14,14))
plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(img_hflip)
```

```python
# 垂直翻轉 (vertical)
img_vflip = img[::-1,:,:]
cv2.imshow('img_vflip', img_vflip)
plt.figure(figsize=(14,14))
plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(img_vflip)
```

```python
# 水平 + 垂直翻轉
img_hvflip = img[::-1,::-1,:]
cv2.imshow('img_hvflip', img_hvflip)
plt.figure(figsize=(14,14))
plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(img_hvflip)
```

```python
# 組合 + 顯示圖片
hflip = np.hstack((img, img_hflip))
vflip = np.hstack((img_vflip, img_hvflip))
img_flip = np.vstack((hflip,vflip))
plt.figure(figsize=(14,14))
plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(img_flip)
```
